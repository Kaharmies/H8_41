## Uniform Distribution

```python
n = 1000
start = 10
width = 20
rv_uniform = uniform.rvs(size=n, loc=start, scale=width)
```

```python
ax = sns.distplot(rv_uniform,
                  bins=100,
                  kde=True,
                  color='skyblue',
                  hist_kws={"linewidth":15, "alpha":1}
                 )
ax.set(xlabel="Uniform Distribution", ylabel="Uniform")
```

```python
n = 100
avg = 0
stdev = 1
rv_normal = norm.rvs(size=n, loc=avg, scale=stdev)
rv_normal
```

```python
ax = sns.distplot(
    rv_normal,
    bins=10,
    kde=True,
    color='skyblue',
    hist_kws={"linewidth": 15, "alpha": 1}
)
plt.xlabel("Normal Distribution") 
plt.ylabel("Frequency")
plt.show()
```

## Gamma Distribution

```python
n = 100
gamma_a = 10
gamma_a = gamma.rvs(a=5, size=1000)
gamma_a
```

```python
ax_gamma = sns.distplot(
    gamma_a,
    bins=100,
    kde=True,
    color='skyblue',
    hist_kws={"linewidth": 15, "alpha": 1}
)
plt.xlabel("Gamma Distribution") 
plt.ylabel("Frequency")
plt.show()
```

## Exponential Distribution

```python
n = 100
loc = 0
scale = 1
rv_exponent = expon.rvs(size=n, scale=scale, loc=loc)
```

```python
ax_expon = sns.distplot(
    rv_exponent,
    bins=100,
    kde=True,
    color='skyblue',
    hist_kws={"linewidth": 15, "alpha": 1}
)
plt.xlabel("Exponentioal Distribution") 
plt.ylabel("Frequency")
plt.show()
```

## Poisson Distribution

```python
from scipy.stats import poisson
```

```python
n = 100
mu = 3
rv_poisson = poisson.rvs(size=n, mu=mu)
```

```python
ax_poisson =  sns.distplot(
    rv_poisson,
    bins=100,
    kde=True,
    color='skyblue',
    hist_kws={"linewidth": 15, "alpha": 1}
)
plt.xlabel("Poisson Distribution") 
plt.ylabel("Frequency")
plt.show()
```

## Binomial Distribution

```python
from scipy.stats import binom
```

```python
rv_binom = binom.rvs(n=10, p=.2, size=100)
rv_binom
```

```python
ax_binom = sns.distplot(
    rv_binom,
    bins=100,
    kde=True,
    color='skyblue',
    hist_kws={"linewidth": 15, "alpha": 1}
)
plt.xlabel("Binomial Distribution") 
plt.ylabel("Frequency")
plt.show()
```
