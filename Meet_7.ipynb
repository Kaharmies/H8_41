```python
years = list(range(1980, 2014))
year80s = list(range(1980, 1990))
year90s = list(range(1990, 2000))
year00s = list(range(2000, 2010))
```

```python
haiti = df_canada.loc[df_canada["Country"] == "Haiti", years].transpose()
haiti.rename(columns={75: "Haiti"}, inplace=True)
haiti.head()
```

```python
haiti.plot()
```

```python
haiti.plot()

plt.ylabel("Number of Immigration from Haiti")
plt.xlabel("Years")
plt.title("Immigration from Haiti per Years")

#syntax plt.text(x, y, label)
plt.text(1999, 6000, "2010 Earthquake")

plt.show()
```

```python
cn_in = df_canada.loc[
    df_canada["Continent"].isin(["China", "India"]),
    years].transpose()
cn_in.rename(columns={
        36: "China",
        79: "India"
    },inplace=True)
cn_in.head()
```

```python
cn_in.plot()
```

```python
top5_countries = df_canada.sort_values(
    by="Total", ascending=False, axis=0
).head()
top5_countries = top5_countries.set_index("Country")
top5_countries = top5_countries[years].transpose()
```

```python
top5_countries
```

```python
top5_countries.plot(figsize=(15, 9))
plt.title("Top 5 Immigrant Origin over Years in Canada")
plt.ylabel("Num of Immigrants")
plt.xlabel("Years")

plt.show()
```

```python
top5_countries.plot(
    kind="area",
    stacked=False,
    figsize=(20, 10)
)

plt.title("Top 5 Immigrant Origin over Years in Canada")
plt.ylabel("Num of Immigrants")
plt.xlabel("Years")

plt.show()
```

```python
least5_countries = df_canada.sort_values(
    by="Total", ascending=False, axis=0
).tail()
least5_countries = least5_countries.set_index("Country")
least5_countries = least5_countries[years].transpose()
```

```python
least5_countries
```

```python
least5_countries.plot(figsize=(15, 9))
plt.title("Top 5 Immigrant Origin over Years in Canada")
plt.ylabel("Num of Immigrants")
plt.xlabel("Years")

plt.show()
```

```python
least5_countries.plot(
    kind="area",
    alpha = .25,
    stacked=True,
    figsize=(20, 10)
)

plt.title("Top 5 Immigrant Origin over Years in Canada")
plt.ylabel("Num of Immigrants")
plt.xlabel("Years")

plt.show()
```

```python
df_iceland = df_canada_country.loc["Iceland", years]
```

```python
df_iceland.plot(
    kind="bar"
)

plt.title("Immigrants from Iceland")
plt.ylabel("Num of Immigrants")
plt.xlabel("Years")

plt.show()
```

```python
df_iceland.plot(
    kind="bar",
    figsize=(15, 5)
)

plt.title("Immigrants from Iceland")
plt.ylabel("Num of Immigrants")
plt.xlabel("Years")

plt.annotate(
    "x", xy=(32, 70), xytext=(28, 20),
    xycoords='data',
    arrowprops=dict(
        arrowstyle="->", connectionstyle="arc3",
        color="blue", lw=2
    )
)

plt.show()
```

```python
top15_countries = df_canada.sort_values(
    by="Total", ascending=False, axis=0
).head(15)
top15_countries = top15_countries.set_index("Country")
top15_countries = top15_countries["Total"]
top15_countries.head(15)
```

```python
top15_countries.plot(
    kind="barh",
    figsize=(15, 15)
)
plt.legend()
plt.show()
```

```python
df_canada_country.groupby("Continent").sum()["Total"].plot(
    kind="pie",
    figsize=(10,10),
    autopct="%1.lf%%",
    startangle=90,
    shadow=True,
    labels=None,
    pctdistance=1.12,
    explode=[.1, 0, 0, 0, 0.1, 0.1]
)
plt.legend(
    labels=df_canada_country["Continent"].values.tolist(),
    loc="upper left",
    fontsize="small"
)

plt.show()
```
