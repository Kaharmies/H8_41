## Regression

```python
import numpy as np
import matplotlib.pyplot as plt
import statsmodels.api as sm
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn import datasets
```

```python
!pip install scikit-learn
```

```python
x = np.array([5, 15, 25, 35, 45, 55]).reshape((-1, 1))
y = np.array([5, 20, 14, 32, 22, 38])

print(x)
print(y)
```

```python
plt.plot(x, y)
plt.scatter(x, y)
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```

```python
#define model
model = LinearRegression()
```

```python
#train linear regression model
model.fit(x, y)
```

```python
model.coef_, model.intercept_
```

```python
lr_score = model.score(x, y)
print("R-Squared:", lr_score)
```

```python
y_hat = model.predict(x)
print("Prediction :", y_hat)
```

```python
plt.plot(x, y, label="target")
plt.scatter(x, y, c="c")

plt.plot(x, y_hat, label="prediction", c="r")
plt.scatter(x, y, c="y")

plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.show()
```

```python
print(model.coef_, model.intercept_)
```

```python
model.predict([[1]])
```

## Multiple Linear Regression

```python
x = np.array([
    [0, 1],
    [5, 1],
    [15, 2],
    [25, 5],
    [35, 11],
    [45, 15],
    [55, 34],
    [60, 35]
])
y = np.array([4, 5, 20, 14, 32, 22, 38, 43])

print(x)
print(y)
```

```python
model = LinearRegression()
```

```python
model.fit(x, y)
```

```python
lr_score_1 = model.score(x, y)
print("R-Squared :",lr_score_1)
```

```python
y_hat1 = model.predict(x)
print("Prediction :", y_hat1)
```

```python
print(model.coef_, model.intercept_)
```

```python
plt.plot(x[:, 0], y, label="target")
plt.plot(x[:, 0], y_hat1, label="prediction", c="r")
plt.scatter(x[:, 0], y, c="y")

plt.xlabel("x1")
plt.ylabel("y1")
plt.legend()
plt.show()

plt.plot(x[:, 1], y, label="target")
plt.plot(x[:, 1], y_hat1, label="prediction", c="r")
plt.scatter(x[:, 1], y, c="y")

plt.xlabel("x2")
plt.ylabel("y2")
plt.legend()
plt.show()
```

```python
print("model weights :", model.coef_)
print("model bias :", model.intercept_)
```

## Linear Regression with Statmodels

```python
x = np.array([
    [0, 1],
    [5, 1],
    [15, 2],
    [25, 5],
    [35, 11],
    [45, 15],
    [55, 34],
    [60, 35]
])
y = np.array([4, 5, 20, 14, 32, 22, 38, 43])

print(x)
print(y)
```

```python
x = sm.add_constant(x)
print(x)
print(y)
```

```python
model = sm.OLS(y, x)
model
```

```python
results = model.fit()
```

```python
print(results.summary())
```
